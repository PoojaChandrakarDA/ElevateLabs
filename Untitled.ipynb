```python
#Task1 identifying missing values, handling them, remove duplicates and inconsistent formatting.
import pandas as pd
# check if panda is install
pd.__version__
# check if it is able to read file
df = pd.read_csv("netflix_titles.csv")
df.head(3)
```

```python
# Handling missing valuue with unknown, not given, not rated
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Not Given")
df['country'] = df['country'].fillna("Unknown")
df['rating'] = df['rating'].fillna("Not Rated")
df['duration'] = df['duration'].fillna("Unknown")
df.head(3)
```

```python
# Formatting -- Capitalizing first letter of column
df.columns = df.columns.str.capitalize()
df.head(3)
```

```python
# Checking if their is any duplicated row

duplicates_count = df.duplicated().sum()
print("Duplicate rows:", duplicates_count)
```

```python
# Making Country column consistent 
Country_rename = {"United States of America":"United States",
                  "USA":"United States",
                  "U.S.A":"United States",
                  "UK":"United Kingdom"
                 }
df['Country'] = df['Country'].replace(Country_rename)
df['Country'] = df['Country'].str.title()
df.head(2)
```

```python
#Make date column consistent from 25 september 2021,25-09-2021, 25/09/2021 
df["Date_added"] = pd.to_datetime(df["Date_added"], errors="coerce", dayfirst=True)
df["Date_added_str"] = df["Date_added"].dt.strftime("%d-%m-%Y")
df.head(20)
```

```python
# Save to a new CSV (recommended)
df.to_csv("netflix_titles_cleaned.csv", index=False)
```
